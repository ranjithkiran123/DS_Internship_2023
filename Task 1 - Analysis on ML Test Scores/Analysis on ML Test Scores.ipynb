<h3><b>Cleaning the column names

```python
df.columns = df.columns.str.replace(' ', '')
df['Score'] = df['Score'].apply(lambda x : (int(x[0])/int(x[4]))*100)
```

```python
df.rename(columns = {'Batch ': 'Batch', 'User_ID ':'User_ID', '   Score   ':'Score'}, inplace= True)
df.columns
```
